```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

%matplotlib inline
sns.set_style("whitegrid")
```

```python
df = pd.read_csv('./data/train_and_test2.csv')
df.head()
```

```python
print("Dataframe shape")
print(df.shape)
print("\n")

print("Column Names: ")
print(df.columns.tolist())
print("\n")

print("Data Info: ")
df.info()
```

```python
print("Missing values: ")
print(df.isnull().sum())
```

```python
df.describe()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

sns.histplot(data=df, x='Fare', ax=axes[0])
axes[0].set_title('Distribution of Passenger Fares (Box Plot)')

sns.boxplot(data=df, x='Fare', ax=axes[1])
axes[1].set_title('Distribution of Passenger Fares (Box Plot)')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Sex', hue='2urvived')
plt.title('Survival Count by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')

plt.legend(title='Survived', labels=['No', 'Yes'])
plt.savefig('images/survival_by_sex.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
survival_rates = df.groupby('Sex')['2urvived'].mean() * 100
print("Survival Rate by Sex (%): ")
print(survival_rates)
```

```python
cross_tab = pd.crosstab(df['Sex'], df['2urvived'])
print("Exact Counts: ")
print(cross_tab)
print("\n")

print("Total Passengers by Sex: ")
print(df['Sex'].value_counts())
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='Sex', y='2urvived', estimator='mean')
plt.title('Survival rate by sex')
plt.xlabel('sex')
plt.ylabel('Survial rate (proportion)')

plt.gca().set_yticklabels(['{:.0f}%'.format(x*100) for x in plt.gca().get_yticks()])
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Pclass', y='Age', hue='2urvived')
plt.title('Distribution of Age by Passenger Class and Survival')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
g = sns.FacetGrid(df, col='Pclass', hue='2urvived', height=4, aspect=1.2, palette='Set1')
g.map(sns.histplot, 'Age', alpha=0.6, bins=20)
g.add_legend(title='Survived', labels=['No', 'Yes'])
plt.savefig('images/age_by_class_survival.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Create a point plot to show the average survival rate for each class, across age, and split by sex.
plt.figure(figsize=(10, 6))
sns.pointplot(data=df, x='Pclass', y='2urvived', hue='Sex', palette='Set1', capsize=0.05, linestyles='-')
plt.title('Survival Rate by Passenger Class and Sex')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate (Proportion)')
plt.gca().set_yticklabels(['{:.0f}%'.format(x*100) for x in plt.gca().get_yticks()])
plt.legend(title='Sex')
plt.savefig('images/survival_by_class_sex.png', dpi=300, bbox_inches='tight')
plt.show()
```
